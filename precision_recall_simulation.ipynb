```python
import matplotlib.pyplot as plt
import numpy as np

def simulation_plot():

    x = np.linspace(0, 5000, 400)
    y = 1000 / (1000 + x)
    
    plt.figure(figsize=(10, 7))
    plt.plot(x, y, color='#40B0C0')

    plt.xlabel('FP / FN count')
    plt.ylabel('Precision / Recall')
    plt.grid(False)

    plt.show()

simulation_plot()
```
